# Build single model to examine splits

```r
train <- expiry_train_df_1 %>% mutate_if(is.character,as.factor)
str(train)
```

```r
test <- expiry_test_df_1 %>% mutate_if(is.character,as.factor)
str(test)
```

```r
W = ifelse(train$renewal_status=="Renewed", 4, 3)

test_tree <- rpart(renewal_status ~ tld + reseller + reseller_country + 
                            region + reg_period + sld_type + sld_length + day_domains + 
                            gibb_score + pattern_domain_count + reg_arpt ,
                   data = train, weights=W,
                   control = rpart.control(maxdepth=10))
```

```r
test_pred <- predict(test_tree, test)
```

```r
test_pred_df = data.frame("actual" = test$renewal_status,
          "predicted" = as.data.frame(test_pred)$Renewed)
head(test_pred_df)
```

```r
lift_df <- chart_lift(pred_df = test_pred_df,
                        dep_var = "actual",
                        pred_var = "predicted")
gains_plot <- plot_gains (lift_df) 
auc <- calc_auc(lift_df)
```
