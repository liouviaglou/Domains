## Feature Eng

```r
count_pct <- function(df) {
  return(
    df %>%
      tally %>% 
      mutate(n_prp = n/sum(n))
  )
}
tld_reseller_names = expiry_train_df_1 %>% 
    group_by(tld_registrar_index) %>% 
    count_pct %>%
    arrange(desc(n)) %>% 
   pull(tld_registrar_index)

metrics_df = metrics_df %>% 
    mutate(tld_reseller = tld_reseller_names)
```

# CREATE new multi-class dependent variables

```r
(auc_vars = grep('auc', names(metrics_df), value=TRUE))
(l10_vars = grep('l10', names(metrics_df), value=TRUE))
```

## Learn to predict renewals by tld-registrar

```r
library(nnet)
model_log <- multinom(renewal_status~tld+registrar, data=expiry_train_df_1)
```

```r
# train and test have diff levels

levels(factor(expiry_train_df_1$registrar))   
levels(factor(expiry_test_df_1$registrar))
```

```r
expiry_test_df_1$registrar <- factor(expiry_test_df_1$registrar, levels=levels(factor(expiry_train_df_1$registrar)))
```

```r
expiry_test_df_1$reseller <- factor(expiry_test_df_1$reseller, levels=levels(factor(expiry_train_df_1$reseller)))
```

```r
# performance on test
predicted <- predict(model_log,newdata = expiry_test_df_1,
                     type = "probs",na.action =na.pass)
predicted_class <- predict(model_log,newdata = expiry_test_df_1,na.action =na.pass)
length(predicted)
head(predicted)
length(predicted_class)
head(predicted_class)

caret::confusionMatrix(as.factor(predicted_class),as.factor(expiry_test_df_1$renewal_status))
```

```r
coefs <- as.data.frame(coef(model_log2))
names(coefs) <- 'value'
coefs %>% arrange(value) %>% head(10)
```

```r
coefs$odds <- exp(coefs$value)
coefs$prob <- (exp(coefs$value))/(1+exp(coefs$value))
coefs["(Intercept)",]
coefs %>% arrange(value) %>% head(10)
```

For **resellernicenic**, the value= **-7.467242**  corresponds to the log odds of the probability of **resellernicenic** renewing. This corresponds to an odds of exp(-7.467242)=	0.0005715021. Which corresponds with a probability of 0.0005711757.

This means providing all the other metrics (tld) are the same, and flipping the reseller to nicenic, the log-odds of renewing will decrease by 7.467242.

In terms of percentage change, the odds for nicenic getting renewed is over 99% lower than the odds for other resellers getting renewed.
